### Campaign DataFrame
----
**Create a Campaign DataFrame that has the following columns:**
- The "cf_id" column.
- The "contact_id" column.
- The “company_name” column.
- The "blurb" column is renamed as "description."
- The "goal" column.
- The "goal" column is converted to a `float` datatype.
- The "pledged" column is converted to a `float` datatype. 
- The "backers_count" column. 
- The "country" column.
- The "currency" column.
- The "launched_at" column is renamed as "launch_date" and converted to a datetime format. 
- The "deadline" column is renamed as "end_date" and converted to a datetime format.
- The "category_id" with the unique number matching the “category_id” from the category DataFrame. 
- The "subcategory_id" with the unique number matching the “subcategory_id” from the subcategory DataFrame.
- And, create a column that contains the unique four-digit contact ID number from the `contact.xlsx` file.
 

Then export the DataFrame as a `campaign.csv` CSV file.


```python
# Create a copy of the crowdfunding_info_df DataFrame name campaign_df. 
campaign_df = crowdfunding_info_df.copy()
campaign_df.head()
```

```python
# Rename the blurb, launched_at, and deadline columns.
crowd_renamed_df = crowdfunding_info_df.rename(columns={
    'blurb':'description','launched_at':'launch_date','deadline':'end_date'})
crowd_renamed_df.head()
```

```python
# Drop extra columns; use axis = 1
crowd_df = crowd_renamed_df.drop("staff_pick", axis=1)
crowd_df = crowd_df.drop("spotlight", axis=1)
crowd_df.head
```

```python
# Get data types
crowd_df.info()
```

```python
# Convert the goal and pledged columns to a `float` data type.
crowd_df[["goal","pledged"]] = crowd_df[["goal","pledged"]].astype('float')
crowd_df
```

```python
# Check the datatypes again
crowd_df.info()
```

```python
# Format the launched_date and end_date columns to datetime format
from datetime import datetime as dt
crowd_df["launch_date"] = pd.to_datetime(crowd_df["launch_date"], unit='s').dt.strftime('%Y-%m-%d')
crowd_df["end_date"] = pd.to_datetime(crowd_df["end_date"], unit='s').dt.strftime('%Y-%m-%d')
crowd_df
```

```python
# Split the category & sub-category column
crowd_df[["category","subcategory"]] = crowd_df["category & sub-category"].str.split('/', n=1, expand=True)
crowd_df
```

```python
# Drop the category & sub-category column
crowd_clean_df = crowd_df.drop("category & sub-category", axis=1)
crowd_clean_df
```

```python
# Merge the campaign_df with the category_df on the "category" column and 
camp_cat_df = crowd_clean_df.merge(categories_df, on='category', how='right')


camp_cat_df.tail(30)
```

```python
# Merge the campaign_df with the subcategory_df on the "subcategory" column.
camp_cat_subcat_df = camp_cat_df.merge(subcategories_df, on='subcategory', how='right')
camp_cat_subcat_df

camp_cat_subcat_df.tail(30)
```

```python
# Export the DataFrame as a CSV file. 
camp_cat_subcat_df.to_csv("Resources/campaign.csv", index=False)
```
